Theory Questions

1. What is unsupervised learning in the context of machine learning.

Ans : Unsupervised learning is a type of machine learning where the algorithm learns patterns from unlabeled data.
  - No predefined outputs (labels)
  - The model discovers hidden structures or patterns
  - Common tasks:
    - Clustering
    - Dimensionality reduction
    - Association rule mining

  - Example: Grouping customers based on purchasing behavior.

2. How does K-Means clustering algorithm work

ans: K-Means groups data into K clusters based on similarity.

  - Steps:

      - Choose K cluster centroids randomly.
      - Assign each data point to the nearest centroid.
      - Recalculate centroids as the mean of assigned points.
      - Repeat steps 2–3 until centroids stop changing.

    Goal: Minimize within-cluster variance (inertia).

3. Explain the concept of a dendrogram in hierarchical clustering.

ANS: A dendrogram is a tree-like diagram used to visualize hierarchical clustering.

  - It shows:

    - How clusters merge or split  -
    - The distance at which clusters combine  -
    - The height of branches indicates similarity distance between clusters.

4.  What is the main difference between K-Means and Hierarchical Clustering

ans :   

      | Feature        | K-Means                       | Hierarchical Clustering       |
      | -------------- | ----------------------------- | ----------------------------- |
      | Cluster number | Must specify **K beforehand** | No need to predefine clusters |
      | Structure      | Flat clustering               | Tree structure                |
      | Scalability    | Faster for large datasets     | Slower                        |
      | Output         | Cluster assignments           | Dendrogram                    |


5. What are the advantages of DBSCAN over K-Means

  ans : DBSCAN (Density-Based Spatial Clustering of Applications with Noise) has several advantages:

  - Can detect arbitrarily shaped clusters-
  - Handles noise/outliers well-
  - Does not require specifying number of clusters-
  - Works well with uneven cluster sizes

6. When would you use Silhouette Score in clustering

ans : The Silhouette Score is used to evaluate clustering quality.

   - It measures:

      - How similar a point is to its own cluster
      - Compared to other clusters
      - Range: -1 to +1

  - Use cases:
    - Choosing optimal number of clusters
    - Comparing clustering algorithms.

7. What are the limitations of Hierarchical Clustering

  ans :
  - High computational cost for large datasets-
  - Sensitive to noise and outliers-
  - Irreversible merges/splits-
  - Memory intensive

8. Why is feature scaling important in clustering algorithms like K-Means

ans: K-Means relies on distance calculations (usually Euclidean distance).

  - If features have different scales:    
  - Example   
    - Income: 10000–100000   
    - Age: 20–60    
  - Income will dominate distance calculations.   
  - Feature scaling (Standardization/Normalization) ensures all features contribute equally.


9.  How does DBSCAN identify noise points

ANS :  DBSCAN categorizes points as:

  - Core points – enough neighbors within radius  
  - Border points – close to core points  
  - Noise points – not reachable from any cluster

  - Noise points:
    - Have fewer than MinPts neighbors within ε distance

10. Define inertia in the context of K-Means

ANS: Inertia measures how tightly clustered the data points are.

It is the sum of squared distances between points and their cluster centroid.

Lower inertia → better clustering.


11. What is the elbow method in K-Means clustering

ANS: The Elbow Method helps determine the optimal number of clusters (K).

  - Steps:-
  - Run K-Means for different K values.
  - Plot K vs Inertia.
  - Find the "elbow point" where inertia stops decreasing significantly.
  - That point indicates the best K.

12. Describe the concept of "density" in DBSCAN

ANS: In DBSCAN, density refers to the number of points within a specified radius (ε).

A cluster forms when regions have high density of points.

Low-density regions separate clusters.

13. Can hierarchical clustering be used on categorical data

ANS: Yes, but not with standard distance metrics like Euclidean distance.

    - instead, use:

    - Hamming distance  
    - Jaccard similarity  
    - Gower distance

These handle categorical variables better.

14. What does a negative Silhouette Score indicate

ANS: A negative silhouette score means:

- A data point is closer to another cluster than its own.

This indicates:

  - Poor clustering
  - Incorrect cluster assignment

15. Explain the term "linkage criteria" in hierarchical clustering

ANS: Linkage criteria determine how the distance between clusters is calculated.

 - Common types:

    - Single linkage – minimum distancE
    - Complete linkage – maximum distance
    - Average linkage – average distance
    - Ward linkage – minimize variance

16. Why might K-Means clustering perform poorly on data with varying cluster  sizes or densities.

ANS:  K-Means assumes:

- Clusters are spherical
- Clusters have similar sizes
- Uniform density

If clusters have:

- Different sizes
- Different densities
- Non-spherical shapes

K-Means may misclassify points.

17. What are the core parameters in DBSCAN, and how do they influence clustering

ANS:

Two main parameters:

1. ε (Epsilon)

Radius of neighborhood around a point.

2. MinPts

ANS : Minimum number of points required to form a dense region.

Influence:

- Small ε → many noise points
- Large ε → clusters merge
- Large MinPts → stricter cluster formation

18. How does K-Means++ improve upon standard K-Means initialization

ANS:  K-Means++ improves centroid initialization.

Instead of random selection:

- First centroid chosen randomly
- Next centroids chosen far from existing ones

Benefits:

  - Faster convergence
  - Better clustering results
  - Reduced chance of poor initialization

19. What is agglomerative clustering

Agglomerative clustering is a bottom-up hierarchical clustering method.

Steps:

- Start with each data point as its own cluster
- Merge the two closest clusters
- Repeat until all points form one cluster
- Result visualized using a dendrogram.

20. What makes Silhouette Score a better metric than just inertia for model evaluation?

ans:  

| Metric             | Inertia          | Silhouette Score         |
| ------------------ | ---------------- | ------------------------ |
| Measures           | Compactness only | Compactness + separation |
| Range              | 0 → ∞            | -1 → 1                   |
| Cluster comparison | Limited          | More reliable            |

```python
# 29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart
# Import libraries
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Step 1: Generate synthetic data
X, y_true = make_blobs(
    n_samples=500,
    centers=4,
    cluster_std=1.2,
    random_state=42
)

# Step 2: Evaluate silhouette scores for k = 2 to 5
k_values = [2, 3, 4, 5]
silhouette_scores = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    labels = kmeans.fit_predict(X)

    score = silhouette_score(X, labels)
    silhouette_scores.append(score)

# Step 3: Plot bar chart
plt.figure(figsize=(8,6))
plt.bar(k_values, silhouette_scores)
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score vs Number of Clusters")
plt.xticks(k_values)
plt.show()
```

```python
# 30. Load the Iris dataset and use hierarchical clustering to group data. Plot a dendrogram with average linkage
# Import libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage

# Step 1: Load Iris dataset
iris = load_iris()
X = iris.data

# Step 2: Compute linkage matrix using average linkage
linked = linkage(X, method='average')

# Step 3: Plot dendrogram
plt.figure(figsize=(10, 6))
dendrogram(linked,
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=False)

plt.title("Hierarchical Clustering Dendrogram (Average Linkage)")
plt.xlabel("Data Points")
plt.ylabel("Euclidean Distance")
plt.show()
```

```python
# 31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with
# decision boundaries
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Step 1: Generate overlapping synthetic data
X, y_true = make_blobs(
    n_samples=500,
    centers=3,
    cluster_std=2.5,   # high std → overlapping clusters
    random_state=42
)

# Step 2: Apply K-Means
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
kmeans.fit(X)

# Step 3: Create a mesh grid for decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(
    np.arange(x_min, x_max, 0.02),
    np.arange(y_min, y_max, 0.02)
)

# Predict cluster for each point in grid
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Step 4: Plot decision boundaries
plt.figure(figsize=(8,6))

# Background (decision regions)
plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')

# Data points
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis', s=50)

# Centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', s=200, marker='X', label='Centroids')

plt.title("K-Means Decision Boundaries on Overlapping Clusters")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

```python

# 34. Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a
# line plot
# Import libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Step 1: Load dataset
data = load_breast_cancer()
X = data.data

# Step 2: Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Compute inertia for K = 2 to 6
k_values = range(2, 7)
inertia_values = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    inertia_values.append(kmeans.inertia_)

# Step 4: Plot inertia vs K
plt.figure(figsize=(8,6))
plt.plot(k_values, inertia_values, marker='o', linestyle='--', color='blue')
plt.title("K-Means Inertia vs Number of Clusters (Breast Cancer Dataset)")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia (Sum of Squared Distances)")
plt.xticks(k_values)
plt.grid(True)
plt.show()
```

```python

# 43 Generate noisy circular data using make_circles and visualize clustering results from KMeans and DBSCAN
# side-by-side
# Import libraries
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN

# Step 1: Generate noisy circular data
X, y_true = make_circles(
    n_samples=500,
    factor=0.5,  # inner circle radius ratio
    noise=0.05,
    random_state=42
)

# Step 2: Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Apply K-Means clustering
kmeans = KMeans(n_clusters=2, init='k-means++', random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)

# Step 4: Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)

# Step 5: Visualize side-by-side
fig, axes = plt.subplots(1, 2, figsize=(14,6))

# K-Means plot
axes[0].scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans_labels, cmap='viridis', s=50, alpha=0.7)
axes[0].scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
                c='red', s=200, marker='X', label='Centroids')
axes[0].set_title("K-Means Clustering")
axes[0].legend()

# DBSCAN plot
axes[1].scatter(X_scaled[:, 0], X_scaled[:, 1], c=dbscan_labels, cmap='viridis', s=50, alpha=0.7)
axes[1].set_title("DBSCAN Clustering")

plt.suptitle("Clustering Comparison: K-Means vs DBSCAN on Noisy Circles", fontsize=16)
plt.show()
```

```python

# 44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.cm as cm

# Step 1: Load Iris dataset
iris = load_iris()
X = iris.data

# Step 2: Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Apply K-Means clustering
k = 3
kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
cluster_labels = kmeans.fit_predict(X_scaled)

# Step 4: Compute silhouette scores for each sample
sample_silhouette_values = silhouette_samples(X_scaled, cluster_labels)
overall_score = silhouette_score(X_scaled, cluster_labels)
print(f"Overall Silhouette Score: {overall_score:.4f}")

# Step 5: Create silhouette plot
fig, ax = plt.subplots(figsize=(8,6))
y_lower = 10

for i in range(k):
    # Aggregate silhouette scores for samples in cluster i
    ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
    ith_cluster_silhouette_values.sort()

    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i

    color = cm.nipy_spectral(float(i) / k)
    ax.fill_betweenx(np.arange(y_lower, y_upper),
                     0, ith_cluster_silhouette_values,
                     facecolor=color, edgecolor=color, alpha=0.7)

    # Label the silhouette plots with cluster numbers
    ax.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
    y_lower = y_upper + 10  # 10 for spacing between clusters

ax.set_title("Silhouette Plot for K-Means Clustering on Iris Dataset")
ax.set_xlabel("Silhouette Coefficient Values")
ax.set_ylabel("Cluster Label")
ax.axvline(x=overall_score, color="red", linestyle="--", label=f"Average Score: {overall_score:.2f}")
ax.set_yticks([])
ax.set_xlim([-0.1, 1])
ax.legend()
plt.show()
```
